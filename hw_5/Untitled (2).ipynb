### Задание 1

Выбран белок геликаза DDX3X 
Функции:
РНК-хеликаза – расплетает вторичную структуру РНК

~Участвует в трансляции (инициации, особенно для мРНК со сложной 5'-UTR)

~Регулирует сплайсинг и экспорт мРНК из ядра

~Важен для врожденного иммунного ответа (распознает вирусную РНК)

~Связан с развитием нейронов (мутации вызывают умственную отсталость)

Особенности:

Локализация: ядро + цитоплазма

Мутации в DDX3X связаны с:

Раком (медуллобластома, лейкемия)

Синдромом DDX3X-ассоциированной интеллектуальной инвалидности

Вирусными инфекциями (ВИЧ, HCV, SARS-CoV-2 эксплуатируют DDX3X)
